<h1 style="text-align: center;">Inflation Forecast Targeting</h1>

## 1. The Central Banker's Problem

Imagine you are a central banker tasked with keeping inflation low and steady at 2\%. You cannot change today's inflation; it is already locked in by past contracts and economic choices. Assume that monetary policy operates with a 1.5-to-2-year "control lag," such that any policy adjustments you make today will only show results down the road. Consequently, you cannot react to economic disruptions in real-time; you can only address them when their impact finally hits years later [1]. Because of this delay, targeting inflation effectively requires you to target an inflation forecast.

## 2. A Structural Macroeconomic Model

Now suppose you and your team develop a structural macroeconomic model governed by the following system of equations:

\begin{align*}
\pi_{t+1} &= \pi_t + \alpha_1 y_t + \alpha_2 x_t + \epsilon_{t+1}\\
y_{t+1} &= \beta_1 y_t - \beta_2(i_t - \pi_t) + \beta_3 x_t + \eta_{t+1}\\
x_{t+1} &= \gamma x_t + \theta_{t+1}
\end{align*}

Where:
* $\pi_t = p_t - p_{t-1}$ represents the inflation rate in year $t$, and $p_t$ is the log price level
* $y_t$ is an endogenous variable, such as the output gap (log output relative to potential output)
* $x_t$ is an exogenous variable
* $i_t$ is the monetary policy instrument (e.g., the repo rate)
* $\epsilon_t$, $\eta_t$, and $\theta_t$ are i.i.d. shocks in year $t$ that are unobserved in year $t-1$

It's assumed the coefficients $\alpha_1$ and $\beta_2$ are strictly positive, while all other coefficients are nonnegative. Additionally, the persistence parameters satisfy $\beta_1 < 1$ and $\gamma < 1$.

## 3. The Policy Objective: An Intertemporal Loss Function

As central bankers, you and your team establish an inflation target, denoted as $\pi^\star$ (e.g., 2% per year). Consequently, the central bank's objective in period $t$ is to choose a path of current and future policy rates $\{i_\tau\}_{\tau=t}^\infty$ that minimizes the intertemporal loss function:

\begin{align}
\mathbb{E}_t \sum_{\tau =t}^\infty \delta^{\tau - t} \mathcal{L}(\pi_\tau) \tag{1}
\end{align}

where $\mathbb{E}_t$ represents the expectation operator conditional on all information available to the central bank at time $t$, and $\delta$ is a discount factor satisfying $0 < \delta < 1$. The period-by-period loss function $\mathcal{L}(\pi_\tau)$ is defined as:

\begin{align}
\mathcal{L}(\pi_\tau) = \frac{1}{2} (\pi_\tau - \pi^\star)^2 \tag{2}
\end{align}

In words, equations (1) and (2) imply that the central bank aims to minimize the expected sum of discounted, squared future deviations of inflation from its target.

## 4. The Two-Year Inflation Forecast and the Optimal Policy Rule

Because the policy rate affects inflation with a two-year lag, it's useful to express future inflation $\pi_{t+2}$ as a function of period-$t$ state variables and the structural shocks from periods $t+1$ and $t+2$. 

Recall, 

\begin{align*}
\pi_{t+1} &= \pi_t + \alpha_1 y_t + \alpha_2 x_t + \epsilon_{t+1} \implies \pi_{t+2} = \pi_{t+1} + \alpha_1 y_{t+1} + \alpha_2 x_{t+1} + \epsilon_{t+2}
\end{align*}

Now we know,
\begin{align*}
\pi_{t+1} &= \pi_t + \alpha_1 y_t + \alpha_2 x_t + \epsilon_{t+1}\\
y_{t+1} &= \beta_1 y_t - \beta_2(i_t - \pi_t) + \beta_3 x_t + \eta_{t+1}\\
x_{t+1} &= \gamma x_t + \theta_{t+1}
\end{align*}

Thus, substiuting in the expression for $\pi_{t+1}, y_{t+1}$ and $x_{t+1}$

\begin{align*}
\pi_{t+2} &= (\pi_t + \alpha_1 y_t + \alpha_2 x_t + \epsilon_{t+1}) + \alpha_1 \left[ \beta_1 y_t - \beta_2 (i_t - \pi_t) + \beta_3 x_t + \eta_{t+1} \right] \\
 &+ \alpha_2(\gamma x_t + \theta_{t+1}) + \epsilon_{t+2} \\
&= a_1 \pi_t + a_2 y_t + a_3 x_t - a_4 i_t + \left( \epsilon_{t+1} + \alpha_1 \eta_{t+1} + \alpha_2 \theta_{t+1} + \epsilon_{t+2} \right)
\end{align*}

where the consolidated coefficients are defined as:

* $a_1 = 1 + \alpha_1 \beta_2$
* $a_2 = \alpha_1 (1 + \beta_1)$
* $a_3 = \alpha_1 \beta_3 + \alpha_2 (1 + \gamma)$
* $a_4 = \alpha_1 \beta_2$

Due to the transmission lag of monetary policy, the policy rate set in period $t$ has no impact on inflation in periods $t$ or $t+1$; its initial effects only materialize in period $t+2$ and onward. By extension, the policy rate established in period $t+1$ remains ineffective until period $t+3$.

In other words, because of this two-year control lag, the central bank's policy choices map to future outcomes in a staggered timeline:
* The repo rate set in year $t$ cannot alter inflation in years $t$ or $t+1$. Its influence is strictly constrained to years $t+2, t+3, \dots$
* Consequently, the repo rate set in year $t+1$ will only begin to affect inflation in years $t+3, t+4, \dots$

Therefore, the central bank can find the optimal repo rate in year $t$ as the solution to the simple period-by period problem
\begin{align}
\min_{i_t} \mathbb{E}_t \delta^2 \mathcal{L} (\pi_{t+2}) = \min_{i_t} \mathbb{E_t}\left[\delta^2 \frac{1}{2} (\pi_{t+2} - \pi^\star)^2\right] \tag{3}
\end{align}

Taking the first-order condition w.r.t. $i_t$:

\begin{align*}
\frac{\partial}{\partial i_t} \mathbb{E}_t \left[ \frac{\delta^2}{2} (\pi_{t+2} - \pi^\star)^2 \right] 
&= \mathbb{E}_t \left[ \delta^2 (\pi_{t+2} - \pi^\star) \frac{\partial \pi_{t+2}}{\partial i_t} \right] \\ 
\end{align*}

Now, 
\begin{align*}
\pi_{t+2} = a_1 \pi_t + a_2 y_t + a_3 x_t - a_4 i_t + \left( \epsilon_{t+1} + \alpha_1 \eta_{t+1} + \alpha_2 \theta_{t+1} + \epsilon_{t+2} \right)
\end{align*}

Thus, 
\begin{align*}
\frac{\partial \pi_{t+2}}{\partial i_t} = \frac{\partial}{\partial i_t}\Big[a_1 \pi_t + a_2 y_t + a_3 x_t - a_4 i_t + \left( \epsilon_{t+1} + \alpha_1 \eta_{t+1} + \alpha_2 \theta_{t+1} + \epsilon_{t+2} \right) \Big] = -a_4 = -\alpha_1\beta_2
\end{align*}

Therefore, 

\begin{align*}
\frac{\partial}{\partial i_t} \mathbb{E}_t \left[ \frac{\delta^2}{2} (\pi_{t+2} - \pi^\star)^2 \right] = \mathbb{E_t} \left[\delta^2 (\pi_{t+2} - \pi^\star)(-a_4) \right] \tag{4}
\end{align*}
Since $\delta^2 > 0$ and $a_4 > 0$ setting (4) to zero we see that:

\begin{align*}
\mathbb{E}_t \left[\delta^2 (\pi_{t+2} - \pi^\star)(-a_4) \right] &= 0 \\ 
\mathbb{E}_t[\pi_{t+2}] &= \pi^\star \\ 
\pi_{t+2 \mid t} &\equiv \mathbb{E}_t[\pi_{t+2}] = \pi^\star \tag{5}
\end{align*}

In words, equation (5) implies that the optimal policy rule requires setting $i_t$ to bring the two-year-ahead expected inflation forecast ($\pi_{t+2 \mid t}$) into exact alignment with the target ($\pi^\star$).

So how do we find $i_t$? well we know:
\begin{align*}
\pi_{t + 2 \mid t} = a_1 \pi_t + a_2 y_t + a_3 x_t - a_4 i_t
\end{align*}
Setting this equal to the target we see:
\begin{align*}
a_1 \pi_t + a_2 y_t + a_3 x_t - a_4 i_t &= \pi^\star \\ 
a_4 i_t &= a_1 \pi_t + a_2 y_t + a_3 x_t - \pi^\star \\ 
i_t &= \frac{a_1 \pi_t + a_2 y_t + a_3 x_t - \pi^\star}{a_4} \\ 
i_t &= \pi_t + b_1(\pi_t - \pi^\star) + b_2 y_t + b_3 x_t
\end{align*}
where 
* $b_1 = \frac{1}{\alpha_1 \beta_2}$
* $b_2 = \frac{1 + \beta_1}{\beta_2}$
* $ b_3 = \frac{\alpha_1 \beta_3 + \alpha_2(1 + \gamma)}{\alpha_1 \beta_2}$

Now, the same result can also be derived using a linear-quadratic regulator, or LQR, where the optimal policy rule is obtained from a Riccati equation.

## 5. Nominal vs. Real Policy Rates

Since central banks change the nominal repo rate ($i_t$), but households and firms do not make decisions based on nominal rates alone. They care about purchasing power and real borrowing costs. In other words, when the central bank raises $i_t$ relative to inflation ($\pi_t$), the real interest rate $r_t$ goes up. This increases the real cost of taking out a mortgage, financing commercial real estate, or investing in new capital. Similarly, the output gap ($y_{t+1}$) creates higher real borrowing costs disincentivize spending, causing economic activity to cool. Therefore we can say, 

\begin{align*}
i_t &= r_t + \pi_t \\ 
r_t &\equiv i_t - \pi_t
\end{align*}

In other words, $r_t$ is the **real policy rate**, or real repo rate. It measures the nominal policy rate adjusted for inflation.

Economically, this matters because households and firms respond to real borrowing costs, not just nominal interest rates. For example, if the nominal interest rate is 5% but inflation is 3%, the real rate is approximately 2%. If inflation rises while the nominal rate stays fixed, the real cost of borrowing falls.

Thus we can write:

$$
y_{t+1} = \beta_1 y_t - \beta_2(i_t-\pi_t) + \beta_3x_t + \eta_{t+1}
$$

Because $\beta_2 > 0$, a higher real policy rate reduces future output:

$$
\frac{\partial y_{t+1}}{\partial (i_t-\pi_t)}=-\beta_2 < 0
$$

Therefore we can say,

$$
y_{t+1} = \beta_1 y_t - \beta_2 r_t + \beta_3 x_t + \eta_{t+1}
$$


## 6. Compact State-Space Form

To solve the central bank's optimization problem cleanly, we group our individual economic equations into a compact state-space model. This framework stacks our separate structural equations into vectors and matrices, allowing us to track the entire macroeconomic system simultaneously.

We define the state vector $s_t$ as deviations from our inflation target $\pi^\star$:

$$
s_t =
\begin{bmatrix}
\pi_t-\pi^\star \\
y_t \\
x_t
\end{bmatrix}
$$

Using our defined real policy control variable ($r_t = i_t - \pi_t$), the dynamic system can be written in matrix form as:

$$
s_{t+1} = A s_t + B r_t + w_{t+1}
$$

where the structural matrices are defined as:

$$
A =
\begin{bmatrix}
1 & \alpha_1 & \alpha_2 \\
0 & \beta_1 & \beta_3 \\
0 & 0 & \gamma
\end{bmatrix}, \qquad 
B =
\begin{bmatrix}
0 \\
-\beta_2 \\
0
\end{bmatrix}, \qquad 
w_{t+1}
=
\begin{bmatrix}
\epsilon_{t+1} \\
\eta_{t+1} \\
\theta_{t+1}
\end{bmatrix}
$$

Going row-by-row we see,

#### Row 1: The Inflation Trajectory
$$
(\pi_{t+1} - \pi^\star) = 1(\pi_t - \pi^\star) + \alpha_1 y_t + \alpha_2 x_t + 0r_t + \epsilon_{t+1}
$$
* **Mechanics:** Tomorrow's inflation gap depends on today's inflation gap, the current output gap ($y_t$), and external shocks ($x_t$).
* **Economic Insight:** The coefficient on the control variable $r_t$ in this row is exactly **$0$**. This provides the formal mathematical proof of the control lag—the central bank's policy choice today cannot directly alter inflation in the next immediate period. 

#### Row 2: The Output Gap Dynamics
$$
y_{t+1} = 0(\pi_t - \pi^\star) + \beta_1 y_t + \beta_3 x_t - \beta_2 r_t + \eta_{t+1}
$$
* **Mechanics:** The future output gap exhibits persistence ($\beta_1$), responds to exogenous pressures ($\beta_3$), and is directly damped by the real policy stance ($-\beta_2$).
* **Economic Insight:** This row represents the transmission channel where the central bank exerts active leverage. By shifting $r_t$, policy alters aggregate demand, which subsequently feeds into the inflation equation during the following period.

#### Row 3: The Exogenous State Variable
$$
x_{t+1} = 0(\pi_t - \pi^\star) + 0 y_t + \gamma x_t + 0r_t + \theta_{t+1}
$$
* **Mechanics:** The variable $x_t$ evolves purely as an autoregressive $\text{AR}(1)$ process with persistence parameter $\gamma$.
* **Economic Insight:** Because $x_t$ represents external economic forces (such as global supply chain disruptions), its trajectory is entirely independent of domestic output or central bank interventions, as denoted by the row's leading zeros.

## 7. The Loss Function in Matrix Form

Recall the period loss is:

$$
\mathcal{L}(\pi_t) = \frac{1}{2}(\pi_t-\pi^\star)^2
$$

Since the first state variable is $\pi_t-\pi^\star$, the loss can be written as:

$$
\mathcal{L}(s_t) = \frac{1}{2}s_t'Q s_t
$$

where

$$
Q =
\begin{bmatrix}
1 & 0 & 0 \\
0 & 0 & 0 \\
0 & 0 & 0
\end{bmatrix}
$$

In this version of the model, the central bank only cares directly about inflation deviations. There is no direct output stabilization term and no direct penalty on interest-rate movements.

Therefore,

$$
R = 0
$$

The infinite-horizon discounted LQR problem is:

$$
\min_{\{r_t\}_{t=0}^{\infty}}
\mathbb{E}_t
\sum_{j=0}^{\infty}
\delta^j
\left[
s_{t+j}'Q s_{t+j}
+
r_{t+j}'Rr_{t+j}
\right]
$$

Since $R=0$, this simplifies to inflation stabilization only:

$$
\min_{\{r_t\}_{t=0}^{\infty}}
\mathbb{E}_t
\sum_{j=0}^{\infty}
\delta^j
s_{t+j}'Q s_{t+j}
$$

## 8. The Bellman Equation and the Riccati Solution

Because the state equation is linear and the loss function is quadratic, this is a linear-quadratic regulator (LQR) problem.

### 8.1 A Brief Explanation of the Bellman Equation

The **Bellman equation** is the central tool of dynamic programming. It restates a
problem of choosing an entire infinite sequence of actions as a recursive
problem of choosing just one action today, correctly.

The key idea is the **principle of optimality**: an optimal plan has the property
that, whatever the initial state and first decision, the remaining decisions must
themselves form an optimal plan for the state that results. This lets us collapse an
infinite-horizon optimization into a single functional equation.

Define the **value function** $V(s)$ as the best (here, lowest) achievable value of
the objective when the system starts in state $s$ and behaves optimally thereafter.
For our discounted control problem, the Bellman equation is:

$$
V(s_t) = \min_{r_t} \Big\{ \underbrace{s_t' Q s_t + r_t' R r_t}_{\text{cost today}}
+ \underbrace{\delta\, \mathbb{E}_t\big[V(s_{t+1})\big]}_{\text{discounted future value}} \Big\}
$$

subject to the law of motion $s_{t+1} = A s_t + B r_t + w_{t+1}$. In words: the value
of being in state $s_t$ equals the cost incurred today plus the discounted expected
value of the state you transition into tomorrow, once you pick the control $r_t$
optimally. The two pieces capture the fundamental trade-off in all dynamic
optimization: a control that looks cheap today may push the system into a costly
future state, and vice versa.

For our linear-quadratic problem we guess $V(s) = s'Ps$. Substituting the guess into
the Bellman equation, carrying out the minimization over $r_t$, and matching terms
yields the discrete-time algebraic Riccati equation for $P$ and the linear
feedback rule $r_t = -Ks_t$ derived below.

### 8.2 A Trivial Worked Example

To see the mechanics without any matrices, consider the simplest possible version:
a one-dimensional system with no shocks. The state is a scalar $s_t$, the control is
a scalar $u_t$, and the dynamics and cost are

$$
s_{t+1} = s_t + u_t, \qquad \text{cost} = \sum_{t=0}^{\infty}\delta^t\big(s_t^2 + u_t^2\big)
$$

Here the "cost today" is $s_t^2 + u_t^2$ (penalizing both distance from the target
state $0$ and the effort $u_t$), and each future period is discounted by $\delta$.
Guess a quadratic value function $V(s) = p\,s^2$ for some unknown constant $p \ge 0$.
The Bellman equation is

$$
p\,s^2 = \min_{u}\Big\{ s^2 + u^2 + \delta\, p\,(s+u)^2 \Big\}
$$

Take the derivative of the bracket with respect to $u$ and set it to zero:

$$
2u + 2\delta p (s+u) = 0 \;\;\Longrightarrow\;\; u^\star = -\,\frac{\delta p}{1+\delta p}\, s
$$

This is the optimal policy: a linear feedback rule $u^\star = -k\,s$ with gain
$k = \dfrac{\delta p}{1+\delta p}$ — exactly the scalar analogue of $r_t = -Ks_t$.
Substituting $u^\star$ back into the Bellman equation gives a single scalar equation
that $p$ must satisfy (the scalar Riccati equation):

$$
p = 1 + \frac{\delta p}{1 + \delta p}
$$

For a concrete number, take $\delta = 0.9$. Solving the quadratic
$0.9\,p^2 - 0.9\,p - 1 = 0$ for the positive root gives $p \approx 1.71$, hence a
feedback gain $k = \dfrac{0.9 \times 1.71}{1 + 0.9 \times 1.71} \approx 0.61$. So the
optimal rule is $u^\star \approx -0.61\, s$: each period you close about 61% of the
gap between the current state and the target, trading off the cost of acting now
against the discounted cost of leaving the system away from target. The full
matrix problem below solves the very same fixed-point logic, only with $P$ and $K$
as matrices instead of scalars.


### 8.3 Solving the LQR Problem via the Riccati Equation

Because the state equation is linear and the loss function is quadratic, this is a linear-quadratic regulator problem.

Guess that the value function is quadratic:

$$
V(s_t)=s_t'Ps_t
$$

The Bellman equation is:

$$
V(s_t) = \min_{r_t} \left\{s_t'Qs_t + r_t'Rr_t + \delta \mathbb{E}_t[V(s_{t+1})]\right\}
$$

Ignoring constants from the shocks, substitute:

$$
s_{t+1}=As_t+Br_t
$$

Then:

$$
V(s_t) = \min_{r_t} \left\{s_t'Qs_t+r_t'Rr_t+\delta(As_t+Br_t)'P(As_t+Br_t)\right\}
$$

The discrete-time algebraic Riccati equation is:

$$
P = Q + \delta A'PA - \delta^2 A'PB (R+\delta B'PB)^{-1} B'PA
$$

The optimal feedback rule is:

$$
r_t=-Ks_t
$$

where

$$
K = (R+\delta B'PB)^{-1} \delta B'PA
$$

Since

$$
i_t = \pi_t+r_t
$$

the nominal policy rate is:

$$
i_t = \pi_t - Ks_t
$$

## 9. Connection to the Inflation Forecast Targeting Rule

Earlier, we derived the two-year inflation forecast:

$$
\pi_{t+2|t} = a_1\pi_t+a_2y_t+a_3x_t-a_4i_t
$$

where

\begin{align*}
a_1 &= 1+\alpha_1\beta_2 \\
a_2 &= \alpha_1(1+\beta_1)\\
a_3 &= \alpha_1\beta_3+\alpha_2(1+\gamma)\\
a_4 &= \alpha_1\beta_2
\end{align*}


The forecast-targeting first-order condition is:

$$
\pi_{t+2|t}=\pi^\star
$$

Substituting the forecast equation:

$$
a_1\pi_t+a_2y_t+a_3x_t-a_4i_t=\pi^\star
$$

Solving for $i_t$:

$$
i_t = \frac{a_1\pi_t+a_2y_t+a_3x_t-\pi^\star}{a_4}
$$

This can be rewritten as:

$$
i_t = \pi_t + b_1(\pi_t-\pi^\star) + b_2y_t + b_3x_t
$$

where

\begin{align*}
b_1 &= \frac{1}{\alpha_1\beta_2}\\
b_2 &= \frac{1+\beta_1}{\beta_2} \\
b_3 &= \frac{\alpha_1\beta_3+\alpha_2(1+\gamma)} {\alpha_1\beta_2}
\end{align*}

Since,

$$
r_t=i_t-\pi_t
$$

we have:

$$
r_t = b_1(\pi_t-\pi^\star) + b_2y_t + b_3x_t
$$

But the LQR solution writes the control as:

$$
r_t=-Ks_t
$$

Therefore, the Riccati equation should imply:

$$
-K =
\begin{bmatrix}
b_1 & b_2 & b_3
\end{bmatrix}
$$

So the Riccati solution and the forecast-targeting first-order condition are two ways of expressing the same optimal policy rule.

## 10. Numerical Implementation

The code below solves the discounted LQR problem numerically and confirms that the Riccati-based feedback gains coincide with the closed-form forecast-targeting coefficients $b_1, b_2, b_3$ derived above.

```python
import numpy as np
from scipy.linalg import solve_discrete_are     

def discounted_lqr(A, B, Q, R, delta):
    """
    Solve the discounted infinite-horizon discrete-time LQR problem.

    Objective:
        sum_{t=0}^\infty delta^t (s_t' Q s_t + u_t' R u_t)

    Dynamics:
        s_{t+1} = A s_t + B u_t

    scipy.linalg.solve_discrete_are solves the undiscounted Riccati equation.
    To handle discounting, use the transformed system:

        A_bar = sqrt(delta) A
        B_bar = sqrt(delta) B

    Then solve the standard discrete algebraic Riccati equation.
    """

    A_bar = np.sqrt(delta) * A
    B_bar = np.sqrt(delta) * B

    P = solve_discrete_are(A_bar, B_bar, Q, R)

    K = np.linalg.solve(
        R + B_bar.T @ P @ B_bar,
        B_bar.T @ P @ A_bar
    )

    return P, K     
```

```python
# Structural parameters
alpha1 = 0.50
alpha2 = 0.20
beta1  = 0.60
beta2  = 0.50
beta3  = 0.30
gamma  = 0.70
delta  = 0.95
pi_star = 0.02

# State-space matrices
A = np.array([
    [1.0, alpha1, alpha2],
    [0.0, beta1,  beta3],
    [0.0, 0.0,    gamma]
])

B = np.array([
    [0.0],
    [-beta2],
    [0.0]
])

# Inflation-only loss
Q = np.diag([1.0, 0.0, 0.0])
R = np.array([[0.0]])
```

```python
P, K = discounted_lqr(A, B, Q, R, delta)

print("Riccati matrix P:")
print(P)

print("\nMatrix K:")
print(K)
```

```python
# Closed-form coefficients
# r_t = b1(pi_t - pi*) + b2 y_t + b3 x_t
b_closed_form = np.array([
    1.0 / (alpha1 * beta2),
    (1.0 + beta1) / beta2,
    (alpha1 * beta3 + alpha2 * (1.0 + gamma)) / (alpha1 * beta2)
])

# Riccati-implied coefficients
# LQR gives r_t = -K s_t
# Therefore b = -K

b_riccati = -K.flatten()

print("Closed-form coefficients:")
print(b_closed_form)

print("\nRiccati-implied coefficients:")
print(b_riccati)
```

## 11. Estimating the Coefficients and Applying to Real-World Data

So far the structural coefficients $(\alpha_1, \alpha_2, \beta_1, \beta_2, \beta_3, \gamma)$
have been assumed. To take the model to real data we must estimate them, and then feed
the estimates into the same LQR machinery to recover an implied policy rule. This section
lays out the workflow and demonstrates it on data.

### 11.1 Mapping the model to observable data

Each structural equation corresponds to a familiar empirical object, and each maps to a
standard macroeconomic time series:

| Model variable | Meaning | Real-world proxy (e.g. FRED series) |
|---|---|---|
| $\pi_t$ | Inflation rate | CPI or core PCE inflation (`CPIAUCSL`, `PCEPILFE`) |
| $y_t$ | Output gap | Real GDP vs. CBO potential (`GDPC1`, `GDPPOT`), or an HP/Hamilton filter |
| $x_t$ | Exogenous driver | Oil/commodity prices, supply-chain index, or FX (`DCOILWTICO`) |
| $i_t$ | Policy instrument | Fed funds / policy rate (`FEDFUNDS`) |

### 11.2 Which equations get estimated 

Only the two behavioral equations contain parameters to estimate; the exogenous process is a
simple AR(1):

$$
\underbrace{\pi_{t+1} = \pi_t + \alpha_1 y_t + \alpha_2 x_t + \epsilon_{t+1}}_{\text{Phillips curve}}
\qquad
\underbrace{y_{t+1} = \beta_1 y_t - \beta_2 (i_t - \pi_t) + \beta_3 x_t + \eta_{t+1}}_{\text{IS / output equation}}
\qquad
\underbrace{x_{t+1} = \gamma x_t + \theta_{t+1}}_{\text{AR(1)}}
$$

Now because every right-hand-side variable is dated $t$ while the dependent
variable is dated $t+1$, each equation can in principle be estimated by a regression of the
lead variable on period-$t$ regressors.

### 11.3 Estimation strategies

1. **Equation-by-equation OLS.** Regress $\pi_{t+1}$ on $(\pi_t, y_t, x_t)$ and $y_{t+1}$ on
   $(y_t, r_t, x_t)$, where $r_t = i_t - \pi_t$. 
2. **State-space + Kalman filter.** Because the output gap $y_t$ is unobserved, the cleanest
   approach treats the whole system as a state-space model and estimates parameters by maximum
   likelihood, letting the Kalman filter infer $y_t$ jointly with the
   coefficients.

### 11.4 From estimates to a policy rule

Once $(\hat\alpha_1, \hat\alpha_2, \hat\beta_1, \hat\beta_2, \hat\beta_3, \hat\gamma)$ are in hand,
rebuild the $A$ and $B$ matrices, solve the discounted Riccati equation with `discounted_lqr`, and
read off the implied feedback rule $r_t = -K s_t$. Comparing the model-implied rate path to what
the central bank actually did yields a policy-stance diagnostic.


The code cell below demonstrates the full loop on simulated data (so the true parameters are
known and estimation accuracy can be checked). The application of this framework to empirical data is left as a future exercise.

```python

import numpy as np

rng = np.random.default_rng(0)

#True parameters (unknown in a real application) 
true = dict(alpha1=0.50, alpha2=0.20, beta1=0.60,beta2=0.50, beta3=0.30, gamma=0.70)
pi_star = 0.02

T = 400
pi = np.zeros(T); y = np.zeros(T); x = np.zeros(T); i = np.zeros(T)
pi[0] = 0.02

for t in range(T - 1):
    i[t] = pi[t] + 1.5 * (pi[t] - pi_star) + 0.5 * y[t]
    r_t  = i[t] - pi[t]
    eps  = rng.normal(0, 0.004)
    eta  = rng.normal(0, 0.004)
    th   = rng.normal(0, 0.004)
    pi[t+1] = pi[t] + true['alpha1']*y[t] + true['alpha2']*x[t] + eps
    y[t+1]  = true['beta1']*y[t] - true['beta2']*r_t + true['beta3']*x[t] + eta
    x[t+1]  = true['gamma']*x[t] + th
i[T-1] = pi[T-1] + 1.5*(pi[T-1]-pi_star) + 0.5*y[T-1]

r = i - pi  


def ols(Y, X):
    """Return coefficient vector for Y = X @ b (no intercept)."""
    b, *_ = np.linalg.lstsq(X, Y, rcond=None)
    return b

# Phillips curve: (pi_{t+1} - pi_t) = alpha1*y_t + alpha2*x_t + eps
dpi = pi[1:] - pi[:-1]
X_pc = np.column_stack([y[:-1], x[:-1]])
alpha1_hat, alpha2_hat = ols(dpi, X_pc)

# Output equation: y_{t+1} = beta1*y_t - beta2*r_t + beta3*x_t + eta
X_is = np.column_stack([y[:-1], r[:-1], x[:-1]])
c_is = ols(y[1:], X_is)
beta1_hat, negbeta2_hat, beta3_hat = c_is
beta2_hat = -negbeta2_hat

# AR(1) for x: x_{t+1} = gamma*x_t + theta
gamma_hat = ols(x[1:], x[:-1].reshape(-1, 1))[0]

est = dict(alpha1=alpha1_hat, alpha2=alpha2_hat, beta1=beta1_hat,
           beta2=beta2_hat, beta3=beta3_hat, gamma=gamma_hat)

print("Parameter   True     Estimated")
print("-" * 32)
for k in true:
    print(f"{k:<9}  {true[k]:>6.3f}   {est[k]:>8.3f}")


A_hat = np.array([
    [1.0, est['alpha1'], est['alpha2']],
    [0.0, est['beta1'],  est['beta3']],
    [0.0, 0.0,           est['gamma']]
])
B_hat = np.array([[0.0], [-est['beta2']], [0.0]])


P_hat, K_hat = discounted_lqr(A_hat, B_hat, Q, R, delta)
b_implied = -K_hat.flatten()   # r_t = b1(pi-pi*) + b2 y + b3 x

print("\nImplied policy-rule coefficients (from estimated model):")
print(f"  b1 = {b_implied[0]:.3f},  b2 = {b_implied[1]:.3f},  b3 = {b_implied[2]:.3f}")

b_true = np.array([
    1.0/(true['alpha1']*true['beta2']),
    (1.0+true['beta1'])/true['beta2'],
    (true['alpha1']*true['beta3']+true['alpha2']*(1.0+true['gamma']))/(true['alpha1']*true['beta2'])
])
print("Closed-form coefficients (from TRUE parameters):")
print(f"b1 = {b_true[0]:.3f},  b2 = {b_true[1]:.3f},  b3 = {b_true[2]:.3f}")
```
